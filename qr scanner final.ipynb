```python
#importing openCV library
import cv2 as cv
```

```python
#reading image
image = cv.imread('qr-code.png')
```

```python
#detecting message of qrcodeusing a openCV's built in function
detector = cv.QRCodeDetector() 
message, qrcode_vertices, qrcode_binary = detector.detectAndDecode(image)

#printing message of qrcode
print('message:', message)
```

```python
#real-time qr code detection
import cv2 as cv
capture = cv.VideoCapture(0)
    

while True:
    ret, frame = capture.read()
    detector = cv.QRCodeDetector()
    #getting message and vertices of QR code
    message, qrcode_vertices, qrcode_binary = detector.detectAndDecode(frame)
    #checking that if it detects vertices and qrcode_verticescontains some value than prints information
    if qrcode_vertices is not None:
        print(qrcode_vertices)
        print('Message', message)
        #making a rectangle around the QR code in realtime
        start_point = qrcode_vertices[0][0]
        end_point = qrcode_vertices[0][2]
        font = cv.FONT_HERSHEY_DUPLEX
        cv.rectangle(frame, tuple(start_point.astype(int)), tuple(end_point.astype(int)), (0,255,255), 2)
        #the following line prints the message hidden in QR code on camera screen above the rectangle
        cv.putText(frame, message, tuple(start_point.astype(int)), font, 1, (100, 20, 155), 2)
    cv.imshow('qrcode_window', frame)
    key = cv.waitKey(1)
    if key == 27:
        break

capture.release()
cv.destroyAllWindows()
```
